```python
import requests
import time
import matplotlib.pyplot as plt

longitudinal = []
latitudinal = []

for i in range(100):
    response = requests.get("http://api.open-notify.org/iss-now.json")
    data = response.json()
    
    longitude = float(data['iss_position']['longitude'])
    latitude = float(data['iss_position']['latitude'])
    timestamp = data['timestamp']
    print(longitude)
    print(latitude)
    print(timestamp)
    
    longitudinal.append(longitude)
    latitudinal.append(latitude)
    
    time.sleep(10)
    
plt.plot(longitudinal, latitudinal)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('ISS Position')
plt.show
```
